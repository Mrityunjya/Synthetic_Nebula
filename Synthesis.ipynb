```python
import pandas as pd

df = pd.read_csv('creditcard.csv')
```

```python
df.info()
df.describe()
```

```python
from imblearn.over_sampling import SMOTE

X = df.drop(['Class'], axis=1)
y = df['Class']

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_res)
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_scaled, y_res)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=50, max_depth=10,random_state=42)
model_rf.fit(X_scaled, y_res)
```

```python
import xgboost as xgb

model_xgb = xgb.XGBClassifier(random_state=42)
model_xgb.fit(X_scaled, y_res)
```

```python
from sklearn.neural_network import MLPClassifier

model_nn = MLPClassifier(random_state=42)
model_nn.fit(X_scaled, y_res)
```

```python
from sklearn.metrics import classification_report

y_pred = model_rf.predict(X_scaled)
print(classification_report(y_res, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_res, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```
